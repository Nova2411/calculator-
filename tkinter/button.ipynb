```python
from tkinter import *
from tkinter.ttk import *
window=Tk()
window.title("welcome to MGMU")
window.geometry('600x300')

selected=IntVar()

lbl=Label(window,text="CHOOSE THE DEPARTMENT", font=("Arial Bold",30))
lbl.pack()

rad1=Radiobutton(window,text='CSE',value=1,variable=selected)
rad2=Radiobutton(window,text='ECE',value=2,variable=selected)
rad3=Radiobutton(window,text='ENTC',value=3,variable=selected)
rad4=Radiobutton(window,text='MECHANICAL',value=4,variable=selected)
rad5=Radiobutton(window,text='CIVIL',value=5,variable=selected)
rad6=Radiobutton(window,text='IT',value=6,variable=selected)

def clicked():
    print(selected.get())
    
btn=Button(window,text="Finalize",command=clicked)

btn.pack()
rad1.pack()
rad2.pack()
rad3.pack()
rad4.pack()
rad5.pack()
rad6.pack()

window.mainloop()
```

```python
from tkinter import *
from tkinter import Menu
window=Tk()
window.title("welcome to MGMU")
window.geometry('600x300')

menu=Menu(window)
new_item=Menu(menu,tearoff=0)

new_item.add_command(label='NEW')
new_item.add_separator()

new_item.add_command(label='EDIT')

menu.add_cascade(label='File',menu=new_item)
menu.add_cascade(label='View',menu=new_item)

window.config(menu=menu)
window.mainloop()
```
